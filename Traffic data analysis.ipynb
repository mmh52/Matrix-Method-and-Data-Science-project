```python
import pandas as pd
df = pd.read_csv("/Users/mdmehedihasan/Downloads/US_Accidents_March23 2.csv")
df.head()
```

```python
df.info()
```

```python
fl_data = df[df["State"] == "FL"]

fl_data .head()
```

```python
fl_data = fl_data.drop(["Source", "ID","End_Lat","End_Lng","Astronomical_Twilight","Wind_Direction","Pressure(in)"], axis=1)
fl_data.head()
```

```python
fl_data.info()
```

```python
fl_data = fl_data.drop(["Nautical_Twilight","Civil_Twilight","Timezone","Airport_Code","Weather_Timestamp","Amenity","Sunrise_Sunset"], axis=1)
fl_data.head()
```

```python
fl_data1 = fl_data.copy()
fl_data1["Start_Time"] = pd.to_datetime(fl_data1["Start_Time"], format="mixed", errors="coerce")
```

```python
fl_data1["End_Time"] = pd.to_datetime(fl_data1["End_Time"], format="mixed", errors="coerce")
fl_data1["end_time"] = fl_data1["End_Time"].dt.strftime("%H:%M:%S")
```

```python
fl_data1["Year"] = fl_data1["Start_Time"].dt.year
fl_data1["Month"] = fl_data1["Start_Time"].dt.month
fl_data1["Day"] = fl_data1["Start_Time"].dt.day
fl_data1["Time"] = fl_data1["Start_Time"].dt.strftime("%H:%M:%S")
```

```python
fl_data1.info()
```

```python
fl_data1.head()
```

```python
fl_data1 = fl_data1.rename(columns={"Time": "start_time"})
fl_data1.head()
```

```python
fl_data1.head()
```

```python
fl_data1 = fl_data1.drop(["Start_Time","End_Time"], axis=1)
fl_data1.head()
```

```python
fl_data1.to_csv("florida_accident_data_long_lat.csv", index=False)
```

```python
severity=fl_data1["Severity"].unique()
severity_list={}
severity_list[1]=fl_data1[fl_data1["Severity"]==1].shape[0]
severity_list[2]=fl_data1[fl_data1["Severity"]==2].shape[0]
severity_list[3]=fl_data1[fl_data1["Severity"]==3].shape[0]
severity_list[4]=fl_data1[fl_data1["Severity"]==4].shape[0]

frame=pd.DataFrame(severity_list,index=["Severity"])
display(frame)
```

```python
import matplotlib.pyplot as plt

map_data = fl_data1.sample(frac=0.2, random_state=1)

plt.figure(figsize=(12,6))
plt.scatter(map_data["Start_Lng"],
            map_data["Start_Lat"],
            s=5,
            c=map_data["Severity"])

plt.colorbar(label="Severity")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("US accidents map")

plt.show()
```

```python
map_data = fl_data1.sample(frac=0.2, random_state=1)

plt.figure(figsize=(12,6))

others = map_data[map_data["Severity"] != 4]
plt.scatter(others["Start_Lng"],
            others["Start_Lat"],
            s=5,
            color="gray",
            alpha=0.5,
            label="Severity 1-3")

sev4 = map_data[map_data["Severity"] ==4]
plt.scatter(sev4["Start_Lng"],
            sev4["Start_Lat"],
            s=8,
            color="red",
            alpha=0.8,
            label="Severity 4")

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("US Accidents Map (Highlighted Severity 4 )")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

map_data = fl_data1.sample(frac=0.2, random_state=1)

plt.figure(figsize=(12,6))

# Plot each severity with a different color
colors = {1: "green", 2: "blue", 3: "orange", 4: "red"}

for sev in [1, 2, 3, 4]:
    subset = map_data[map_data["Severity"] == sev]
    plt.scatter(subset["Start_Lng"],
                subset["Start_Lat"],
                s=6,
                color=colors[sev],
                alpha=0.6,
                label=f"Severity {sev}")

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("US Accidents Map (Colored by Severity)")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

county_name = "Lee"

county_data = fl_data1[fl_data1["County"] == county_name]
map_county_data = county_data.sample(frac=0.5, random_state=1)

plt.figure(figsize=(12,6))

colors = {1: "green", 2: "blue", 3: "orange", 4: "red"}

for sev in [1, 2, 3, 4]:
    subset = map_county_data[map_county_data["Severity"] == sev]  # ✅ fixed
    plt.scatter(subset["Start_Lng"],
                subset["Start_Lat"],
                s=6,
                color=colors[sev],
                alpha=0.6,
                label=f"Severity {sev}")

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title(f"{county_name} County Accidents Map")  # ✅ fixed
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

county_name = "Lee"

county_data = fl_data1[fl_data1["County"] == county_name]
map_county_data = county_data.sample(frac=0.7, random_state=1)

plt.figure(figsize=(12,6))

colors = {1: "green", 2: "blue", 3: "orange", 4: "red"}

for sev in [1, 2, 3, 4]:
    subset = map_county_data[map_county_data["Severity"] == sev]
    plt.scatter(subset["Start_Lng"],
                subset["Start_Lat"],
                s=6,
                color=colors[sev],
                alpha=0.6,
                label=f"Severity {sev}")

sev4 = map_county_data[map_county_data["Severity"] == 4].head(20)

for _, row in sev4.iterrows():
    plt.text(row["Start_Lng"],
             row["Start_Lat"],
             row["Street"],
             fontsize=10,
             color="purple", 
             alpha=0.8)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title(f"{county_name} County Accidents Map")
plt.legend()

plt.show()
```

```python
lee_sev4 = fl_data1[(fl_data1["County"] == "Lee") &(fl_data1["Severity"] == 4)]
lee_sev4.head()
```

```python
lee_sev4.info()
```

```python
top_streets = lee_sev4["Street"].value_counts().head(10)

top_streets.plot(kind="bar", figsize=(10,5))
plt.ylabel("Number of Accidents")
plt.title("Top 10 Streets with Most Severity 4 Accidents (Lee County)")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

hb = plt.hexbin(
    lee_sev4["Start_Lng"],
    lee_sev4["Start_Lat"],
    gridsize=70,      
    cmap="hot",       
    mincnt=1,
    linewidths=0.2
)

plt.colorbar(hb, label="Severity 4 Accident Count")


plt.xlim(lee_sev4["Start_Lng"].min() - 0.01, lee_sev4["Start_Lng"].max() + 0.01)
plt.ylim(lee_sev4["Start_Lat"].min() - 0.01, lee_sev4["Start_Lat"].max() + 0.01)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Lee County Severity 4 Accident Hotspots")
plt.grid(alpha=0.2)

plt.show()
```

```python
lee_sev4 = lee_sev4.copy()

lee_sev4["Hour"] = pd.to_datetime(lee_sev4["start_time"], format="%H:%M:%S").dt.hour

lee_sev4["Hour"].value_counts().sort_index().plot(kind="bar", figsize=(10,5))
plt.xlabel("Hour of Day")
plt.ylabel("Number of Accidents")
plt.title("Severity 4 Accidents by Hour in Lee County")
plt.show()
```
